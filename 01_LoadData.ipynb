```python
import pandas as pd
Raw_Data = pd.read_csv('Raw_Data.csv')
Raw_Data.head()
```

```python
df = pd.concat([Raw_Data.iloc[:, 0], Raw_Data.iloc[:, 4], Raw_Data.iloc[:, 5], Raw_Data.iloc[:, 6]], axis = 1)
df.rename(columns = {'Target(Class)' : 'Label'}, inplace = True)
df.head()
```

```python
df1 = pd.DataFrame(df[df['Label'] == 1].iloc[:, 1:4] * 9.81)
df1 = df1.reset_index(drop = True)
df1.head()
```

```python
df2 = pd.DataFrame(df[df['Label'] == 2].iloc[:, 1:4] * 9.81)
df2 = df2.reset_index(drop = True)
df2.head()
```

```python
df3 = pd.DataFrame(df[df['Label'] == 3].iloc[:, 1:4] * 9.81)
df3 = df3.reset_index(drop = True)
df3.head()
```

```python
df4 = pd.DataFrame(df[df['Label'] == 4].iloc[:, 1:4] * 9.81)
df4 = df4.reset_index(drop = True)
df4.head()
```

```python
df1.to_csv('Data/SlicedData/df1.csv', sep = ',')
df2.to_csv('Data/SlicedData/df2.csv', sep = ',')
df3.to_csv('Data/SlicedData/df3.csv', sep = ',')
df4.to_csv('Data/SlicedData/df4.csv', sep = ',')
```
